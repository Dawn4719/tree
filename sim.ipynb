```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch  # 用于创建自定义图例

'sx-mathoverflow'
'sx-superuser'
'wiki-talk-temporal'
'sx-stackoverflow'
'sx-askubuntu'
di = {}
di['sx-mathoverflow'] = "Mathoverflow"
di['sx-superuser'] = "Superuser"
di['sx-stackoverflow'] = "Staclpverflow"
di['sx-askubuntu'] = "AskUbuntu"

for data in ['sx-askubuntu', 'sx-stackoverflow', 'sx-superuser', 'sx-mathoverflow']:
    date = ["1H", "12H", "1D", "3D", "7D", "15D", "1M", "2M", "4M"]
    file_paths = []
    for i in date:
        file_paths.append(data + "_CAP" + i + ".txt")

    # 存储每个数据集的快照大小和交集大小
    snapshot_data = []
    intersection_data = []

    # 逐个读取文件
    for file_path in file_paths:
        with open(file_path, "r") as f:
            lines = f.readlines()
            snapshot_sizes = list(map(int, lines[0].strip().split()))  # 第一行快照大小
            intersection_sizes = list(map(int, lines[1].strip().split()))  # 第二行交集大小
            # 将每个文件的数据添加到列表
            l = []
            for i in range(len(intersection_sizes)):
                l.append((snapshot_sizes[i] + snapshot_sizes[i + 1] - 2 * intersection_sizes[i]) / (snapshot_sizes[i] + snapshot_sizes[i + 1]))
            snapshot_data.append(snapshot_sizes)
            intersection_data.append(l)

    print(len(snapshot_data))
    print(len(intersection_data))

    # 设置绘图
    fig, ax1 = plt.subplots(figsize=(12, 10))

    # 创建右边的第二个纵坐标轴
    ax2 = ax1.twinx()

    # 横坐标位置
    x = np.arange(9) 
    width = 0.4     # 箱线图的宽度

    color1 = "#2e6f9b"
    color2 = "#d77e2c"

    color1 = "#cee6f6" 
    color2 = "#d6c8b7"

    color1 = "#92b9c6"
    color2 = "#989a98"


    # 在右侧纵坐标上绘制箱线图（第一个数组）
    positions_right = x + width / 2 + 0.05
    bp2 = ax2.boxplot(snapshot_data, positions=positions_right, widths=width, patch_artist=True, 
                    boxprops=dict(facecolor=color1, color="black", linewidth=2), 
                    whiskerprops=dict(color="black", linewidth=2),
                    capprops=dict(color="black", linewidth=2),  
                    medianprops=dict(color="black", linewidth=2),
                    showfliers=False,
                    )

    # 在左侧纵坐标上绘制箱线图（第二个数组）
    positions_left = x - width / 2
    bp1 = ax1.boxplot(intersection_data, positions=positions_left, widths=width, patch_artist=True, 
                    boxprops=dict(facecolor=color2, color="black", linewidth=2), 
                    whiskerprops=dict(color="black", linewidth=2),
                    capprops=dict(color="black", linewidth=2),  
                    medianprops=dict(color="black", linewidth=2),
                    showfliers=False,)

    # 设置横坐标刻度
    ax1.set_xticks(x)
    ax1.set_xticklabels(date, rotation=310)
    fontsize = 45
    # 设置纵坐标和标签
    ax1.set_ylabel("Porprotion", fontsize = fontsize)
    ax2.set_ylabel("$\\#$ of edges on Snapshots", fontsize = fontsize)
    ax1.set_xlabel("Snapshots with different time interval", fontsize = fontsize)

    ax1.tick_params(axis='x', labelsize=fontsize)
    ax1.tick_params(axis='y', labelsize=fontsize)
    ax2.tick_params(axis='y', labelsize=fontsize)

    ax1.set_yticks([0.5, 0.6, 0.7, 0.8, 0.9, 1])
    # 右侧纵坐标刻度（对数刻度）
    ax2.set_yscale('log')  # 保留对数刻度
    if data == "sx-stackoverflow":
        ax2.set_ylim(1, 100000000)
    else:
        ax2.set_ylim(1, 1000000)

    ax2.set_yscale('log')

    # 添加图例
    ax1.legend([bp2["boxes"][0], bp1["boxes"][0]], ["$\\#$ of edges on Snapshots", "Proportion"], loc="lower right", fontsize=20)
    # ax2.legend([], [], loc="lower right", bbox_to_anchor=(0.75, 0.05))

    # plt.title(data, fontsize=fontsize)
    plt.tight_layout()
    plt.savefig(data+".pdf", transparent=True, bbox_inches = 'tight')
    plt.show()
```
